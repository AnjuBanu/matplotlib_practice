```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import math
%matplotlib inline
```

```python
x = np.arange(0, math.pi*2, 0.05)
x
```

```python
y = np.sin(x)
y
```

```python
plt.xlabel("angle")
plt.ylabel("sine")
plt.title("sine plot")
plt.plot(x,y)
```
