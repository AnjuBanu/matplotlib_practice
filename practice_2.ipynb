```python
%config Completer.use_jedi = False
import matplotlib.axes
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.set_title("Sine wave")
ax.set_xlabel("Angle")
ax.set_ylabel("Sine")
x=np.linspace(-3,3,100)
y = np.sin(x)
ax.plot(x,y)
```

```python
fig2 = plt.figure(edgecolor="r",figsize=(5,5),frameon=False,facecolor="g")
ax2 = fig2.add_axes([0,0,1,1])
ax2.set_title("Sine wave")
ax2.set_xlabel("Angle")
ax2.set_ylabel("Sine")
x=np.linspace(-3,3,100)
y = np.sin(x)
ax2.plot(x,y)
ax2.plot(x,np.cos(x))
ax2.legend(labels = ('sin','cos'), loc=0)
```
